# Введение

**Базовый глоссарий маркетолога**
* **Веб-аналитика** – сбор, измерение, представление, количественный и качественный анализ информации о посетителях интернет-ресурсов (их поведение на сайтах). Получаемые данные используются для улучшения и оптимизации сайтов с целью расширения их функциональных возможностей, а также для более эффективного расходования рекламных бюджетов.
* **Интернет-канал** – ресурс или сервис источника рекламы, с которого приходит целевая аудитория. Например: источник — Яндекс, а канал — контекстная реклама Яндекс.Директ.
* **Клик** – нажатие пользователем на ссылку или баннер.
* **Коллтрекинг** – мониторинг звонков, совершённых после показа рекламы, в основе которого лежит подмена телефонных номеров.
* **Конверсия** – доля сеансов, в которых пользователи выполнили целевое действие от общего количества сеансов за отчётный период.
* **Лид** (от англ.*lead* – вести) – потенциальный клиент, тем или иным образом отреагировавший на маркетинговую коммуникацию.
* **Охват** – количество человек, проконтактировавших с рекламным сообщением хотя бы один раз в течение рассматриваемого периода.
* **Парсер** – скрипт или программа, используемые для сбора информации с сайтов.
* **Поисковая оптимизация** – комплекс мер для поднятия позиций сайта в результатах выдачи поисковых систем по определённым запросам пользователей. В подобные меры входит: работа с контентом, мета-тегами, заголовками сайта, создание и работе со ссылочной массой.
* **Показы** – появления вашего объявления на страницах, загружаемых пользователями.
* **Рекламная кампания** – комплекс мероприятий, направленных на онлайн-рекламу товаров и услуг компании. Рекламная кампания в интернете состоит из медиапланирования, проведения кампании (с коррекцией плана по мере необходимости) и оценки результатов.
* **Ставка** – цена клика, которую готов заплатить рекламодатель. Ставка может равняться минимальной цене клика, но не может быть меньше её.
* **Таргетинг** – рекламный механизм, с помощью которого из общей аудитории выделяют целевую группу пользователей, соответствующую заданным критериям, и показывают рекламу именно этой группе.
* **Трафик** (англ. *traffic* — движение, транспорт). В контексте интернет-маркетинга означает объём посетителей сайта за отчётный период. Измеряется чаще всего в количестве визитов на сайт.
* **Data-driven** (англ. управляемый данными) – подход к управлению, который основывается на собираемых данных.
* **KPI** – ключевые показатели эффективности, используемые для измерения поставленных перед компанией целей в социальных сетях.
* **LTV (Lifetime Value)** – прибыль, которую приносит пользователь за всё время работы с ним, «жизненный цикл» клиента.
* **ROI (Return On Investment)** – коэффициент рентабельности инвестиций, который выражается в процентах дохода на каждый вложенный рубль. Чем выше показатель, тем выше уровень возврата инвестированных средств. ROI = (доход – сумма инвестиций) ÷ количество вложенных средств × 100%.

**Целевая аудитория** — это объединённая общими признаками группа людей, заинтересованных в товарах или услугах компании. Под общими признаками могут пониматься любые характеристики (не только пол, возраст), необходимые компании для продвижения своего продукта (например, работающие замужние женщины от 25 до 35 лет, носящие очки). Именно на них в первую очередь направлена маркетинговая активность.

**Нужно ли сегментировать аудиторию для эффективной работы рекламы?** \
Да. Когда мы сегментируем аудиторию и подбираем более персонализированное предложение для решения задачи конкретных пользователей, то можем увеличить количество продаж и средний чек. Более того, аудитория, как правило, неоднородна, с разной мотивацией к покупке, а значит, для разных сегментов будут актуальны разные УТП (уникальные торговые предложения).

**Уникальное торговое предложение** — это характеристика товара или бренда, которая выделяет его на фоне конкурентов и на основании которой выстраивается маркетинговая коммуникация. \
никальное торговое предложение (УТП) — это характеристика, позволяющая продукту, услуге или бренду выгодно отличаться от конкурентов. Часто УТП используется для отстройки от конкурентов: подчёркивается собственное преимущество, которым конкурент не обладает.

**Как понять, что интернет-реклама работает хорошо?** \
От рекламы мы ждём новых продаж, но важно, чтобы она работала «в плюс», чтобы доходы от неё покрывали расходы.
Во время проведения рекламных кампаний необходимо отслеживать, сколько мы тратим за один переход по рекламе, за привлечение одного покупателя; не выходят ли эти показатели за предельно допустимые — те, которые мы можем себе позволить. Важно мониторить отмеченные показатели постоянно, не только на старте кампании.
Это позволит вовремя корректировать рекламные настройки и предложения для клиентов.

**Как нам выяснить, много ли пользователей приходит на наш сайт благодаря рекламным кампаниям и каковы источники трафика?** (Мы не знаем, сколько пользователей появляется на сайте каждый день, да ещё и в разбивке по разным кампаниям.) \
Необходимо установить счётчики систем аналитики. 
установка счётчиков Яндекс Метрики и Google Аналитики поможет нам понять, какое количество пользователей приходит на сайт каждый день, откуда они приходят, какие действия совершают.

В системах аналитики мы сможем анализировать данные в разных срезах, таких как источники перехода пользователей, рекламные кампании. Для оценки эффективности работы источников мы настроим нужные цели и будем сравнивать источники друг с другом.
Счётчики аналитики также помогут нам установить, каков пол и возраст нашей целевой аудитории, из каких регионов и городов чаще совершаются покупки. Кроме того, мы сможем проанализировать поведение пользователей и построить воронку продаж.

**Каждое ли действие пользователя на сайте можно отследить?** \
При умелой настройке можно отслеживать практически любое действие пользователя на сайте: посещение разных страниц, добавление товаров в корзину, их оформление и покупку, заполнение и отправку форм и т.п.

**Воронка продаж** — путь, который пользователь проходит до покупки. Каждый шаг, который совершает пользователь на пути к покупке — это шаг в воронке продаж. Воронку важно уметь строить и анализировать, так как её исследование помогает находить точки роста для увеличения продаж, а также прогнозировать нужный объём трафика.

# Цели и задачи веб-аналитики

**Веб-аналитика** - это сбор и последующий анализ данных в интернет-маркетинге. 

**Цель веб-аналитики**: анализ ресурса и каналов привлечения трафика для последующего улучшения самого ресурса и оптимизации каналов трафика. 

**Задачи веб-аналитики**:
* Улучшить ресурс: сайт, мобильное приложение, сервис. Мы смотрим на сайт через призму клиентов. 
* Оценить качество трафика: В первую очередь платного трафика.
* Исключить субъективность: и лучше понять свою аудиторию. 
* Оптимизировать продуктовую линейку или услуги: Больше инсайтов про то, что ЦА готова купить в интернете. 

**Веб-аналитика не спасет, если...:**
* Мало данных. Например, в день приходят 10 пользователей на сайт, по каналу за месяц получили 2 конверсии, за неделю 6 установок приложений. 
* Некорректно оцениваются насыщение конверсии: Если уже все хорошо, то скачка в 2 раза и более не получить.
* Данные не корректны. 
* Нет связи между действиями в интернете и бизнес-показателями: Отдельно считаются клики на рекламу, отдельно продажи. 

# Данные веб-аналитики

**3 состояния данных веб-аналитики**:
1. Мало данных. На этом этапе обычно приходят к осознанию недостаточности информации для принятия решений. Это обычно какой-нибудь стартап или начальная стадия интернет-компании. Начинается корректная настройка сбора информации. 
2. Достаточное количество информации. Рождается множество табличных и дашбордов, и порой много информации собирается бесполезно. Мы начинаем оптимизировать время и состредоточение ресурсов. 
3. Большой объем данных. Data driven. Как правильно собирать, хранить и обрабатывать весь массив данных, чтобы это приносило пользу и окупало затраты. 

**Какие данные мы можем собирать в рамках веб-аналитики:**
* Показы материалов аудитории. Это самый верхний уровень информации
* Объем трафика.
* Понять ЦА (пол, возраст, другие демографические показатели и индикаторы их предпочтений). Нужно для таргетирования
* Источники трафика. 
* Качество трафика. Уровень вовлечения в контент. 
* Целевые действия (заказы, заявки с сайта и пр.)

**Понятия качества трафика**:
1. **Отказы** - доля пользователей, которые закрыли сайт в течение первых 15 секунд, проведенных на нем. Причины: (1) долгая загрузка сайта, (2) сайт не адаптирован под устройство, (3) юзеру не интересен продукт. Оптимальное количество отказов - 20-30%. 
2. **Время на сайте** - показатель, который отмечает, сколько времени пользователи (в среднем) проводят на сайте. Универсального оптимального показателя тут нет. В среднем нужно полторы минуты, чтобы совершить целевое действие. С какого-то канала приходят пользователи со среднем временем на 30 секунд. 
3. **Глубина просмотра** - это среднее количество просмотренных страниц одним пользователем за сеанс. Также нет универсального показателя. 

# Метрики и бизнес-показатели

**Типы метрик**:
1. Данные по сайту
    * Визиты
    * Показатель отказов
    * Глубина просмотров
    * Время на сайте
    * Уровень вовлеченности
    * Новые и вернувшиеся пользователи
2. Данные по конверсии
    * Стоимость конверсии (какое количество рекламных расходов на 1 целевое действие)
    * Доход с каждого клика по рекламе
    * Средняя стоимость продажи с сайта
    * Возвращение затрат на рекламу с конкретного рекламного канала
3. Бизнес-показатели (чем интересуется заказчик, владелец бизнеса)
    * Окупаемость инвестиций на рекламу
    * LTV (доход с клиента за все время взаимодействия с ним). 
    * Средний чек 
    * Коэффициент закрытия лидов (лиды - это тот, который указал свои контактные данные)
    * Коэффициент удержания клиентов
    * Частота покупок
    * Частота повторных покупок

# Квиз

**Как можно обозначать конверсию?**
* **Conversion rate (CR)** — конверсия, отношение числа посетителей сайта, выполнивших на нём какие-либо целевые действия, к общему количеству посетителей сайта.

Но термин часто используется и для обозначения количества пользователей, перешедших с одного этапа воронки на другой. Например, у нас такая воронка: зашли на сайт — посмотрели условия доставки — положили товар в корзину — оформили заказ.Тогда будет конверсия из посетителей сайта — в перешедших на страницу доставки, из перешедших на страницу доставки — в положивших товар в корзину и т. д.

Для удобства в отчётах такие конверсии могут обозначать **C1**, **C2**, **C3** и т. д. (в зависимости от количества этапов в воронке). С0 будет обозначать сквозную конверсию — из первого этапа в последний.

**CTR** — тоже конверсия, но обозначает конкретный этап — отношение количества кликнувших на баннер или рекламное объявление к увидевшим.

Рассчитайте количество кликов по баннеру, чтобы установить, какое количество пользователей перешли на сайт благодаря баннеру.


**Вычислите drop-off rate.** \
**Drop-off rate** — это процент выходов. В данном случае это пользователи, которые кликнули на баннер, но не перешли на сайт, закрыли его.

Причины могут быть разные. Первая — сайт долго грузился: с увеличением загрузки сайта до 5 секунд показатель отказов увеличивается до 90%.
Вторая — пользователи случайно кликнули по рекламе.
Третья — реклама вводит пользователя в заблуждение, например, «крестик», обычно обозначающий возможность закрыть рекламу, оказался элементом картинки, а кнопка «Закрыть» была спрятана.

Деньги при этом списываются за клик, поэтому важно знать drop-off rate.

Процент выходов позволяет нам понимать, есть ли проблемы с сайтом, отвечает ли рекламное объявление ожиданиям пользователя. Так как каждый клик по рекламному объявлению оплачивается из нашего рекламного бюджета, то необходимо следить за работой сайта, отвечает ли объявление запросу целевой аудитории, ведёт ли оно на нужную страницу.


**Какой процент пользователей конвертировался из бесплатного продукта в платный? Какой процент «отвалился»?** \
Free trial — бесплатная версия продукта, бесплатный пробный период в приложении или сервисе: кино, музыка, фитнес, медитации, курсы. После пробного периода часть клиентов продолжает пользоваться продуктом (**retention rate** — коэффициент удержания), часть отказывается (**churn rate** — коэффициент оттока).

Также эти два показателя можно использовать для расчёта экономики в любых подписных моделях бизнеса вне зависимости от наличия пробного периода.

Например, каков коэффициент удержания на третий год подписки по сравнению со вторым. Знание этих показателей позволяет привлекать новых пользователей уже сейчас, чтобы потом не было ситуации, когда старые клиенты перестали пользоваться продуктом, а новых клиентов нет.

ROMI учитывает только маркетинговые затраты, и не включает затраты на производство товара, зарплаты сотрудников, аренду помещения.

**Какая проблема у канала Яндекс — РСЯ?**

Рекламная сеть Яндекса (РСЯ) — это система размещения рекламы на сервисах Яндекса и площадках партнёров.
РСЯ открывает рекламодателям Директа доступ к аудитории более 40 000 площадок: сайтов, мобильных приложений и приложений Smart TV.
Партнёры РСЯ получают доход за размещение рекламы, а рекламодатели — приток трафика и клиентов.

РСЯ работает так: пользователь заходит на сайт, например, Gismeteo.ru, чтобы посмотреть прогноз погоды. В какой-то части этого сайта появляется рекламный баннер.
Модель потребления рекламного контента в этом случае сильно отличается от той, что есть в поисковике того же Яндекса.
В поисковике обычно изначально ищут товар/услугу или информацию, связанную с ними. И рекламные объявления настраивают под эти запросы.

Поэтому в поиске человек больше мотивирован на клик по рекламному объявлению и покупку. А значит сравнивать конверсию и CPC в РСЯ с этими же показателями в поисковом контексте некорректно. Конверсия в клик в РСЯ всегда будет меньше, а управлять CPC можно только через уменьшение CPM.

Так как в этом канале у вас не сходится экономика, нужно решить для себя:

1. Вы работаете на привлечение пользователей и увеличение продаж, а значит или оптимизируете рекламные объявления, чтобы улучшить показатели конверсии и/или CPC, или отключаете рекламу и перераспределяете бюджет на более эффективные каналы;
2. Вы работаете на увеличение охватов и узнавание бренда, а значит исключаете этот канал из отчёта и будете оценивать отдельно по другим метрикам (охватам, top of mind, знанию бренда с подсказкой и без, динамике роста органического трафика).


Конверсия в клик ниже, чем у других каналов, из-за чего стоимость клика значительно выше.
Но здесь важно разобраться, какая цель изначально была у этого рекламного канала. Чаще всего медийные сети (РСЯ и КМС) используют чтобы привлечь холодный трафик или увеличить знание бренда.

Цена за клик выше.

По воронке AIDA такие размещения часто направлены на этап первого
А — Attention. То есть задача этого рекламного канала — работать над «здоровьем бренда». Это предполагает работу с такой информацией и метриками, как знание бренда с подсказкой и без, опыт первой и повторных покупок, лояльность аудитории и готовность рекомендовать бренд, понимание сути бренда и т. д.
